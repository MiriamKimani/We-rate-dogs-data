```python
import json

tweet_json = []

with open('tweet-json.txt', 'r') as file:
    for line in file:
        # Loading line object
        line_data = json.loads(line)
        
        # Extract the columns we want to use
        tweet_id = line_data['id']
        retweet_count = line_data['retweet_count']
        favorite_count = line_data['favorite_count']
        
        
        
        # Put the data extracted into the list we created
        tweet_json.append({'tweet_id': tweet_id, 
                          'retweet_count': retweet_count,
                          'favorite_count': favorite_count
                         })    
```

```python
#create dataframe
tweet_json_df = pd.DataFrame(tweet_json, columns = ['tweet_id', 'retweet_count', 'favorite_count'])      
```

## Cleaning Data

```python
# Make copies of original pieces of data
twitter_archive_copy = twitter_archive.copy()
image_prediction_copy = image_prediction.copy()
tweet_json_df_copy = tweet_json_df.copy()
```

### Issue #1:

Multiple datasets which give related information. 

#### Define:
Merge all the three datasets so that it is easier to work with.

#### Code

```python
twitter_data = pd.concat([twitter_archive_copy, image_prediction_copy, tweet_json_df_copy], join='outer', axis=1)
```

#### Test

```python
twitter_data.columns
```

```python
twitter_data
```

### Issue #2:
Wrong datatype in the twitter_id column.

#### Define
Change twitter_id into an string.

#### Code

```python
twitter_data.tweet_id = twitter_data.tweet_id.astype('str')
```

#### Test

```python
twitter_data.tweet_id.info()
```

### Issue #3:
Timestamp is in the wrong format.

#### Define:
Change timestamp format to datetime.

#### Code:

```python
twitter_data.timestamp
```

```python
#First remove the last 6 characters.
#Then now change into datetime.
twitter_data.timestamp = twitter_data.timestamp.str.slice(start=0, stop=-6)
twitter_data.timestamp = pd.to_datetime(twitter_data.timestamp, format = "%Y-%m-%d %H:%M:%S")
```

#### Test:

```python
twitter_data.timestamp
```

```python
twitter_data.info()
```

### Issue #4:
Unnecessary columns.

#### Define:
Remove the following columns (in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id, retweeted_status_user_id, retweeted_status_timestamp)

#### Code:

```python
twitter_data.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'], axis=1, inplace = True)
```

#### Test:

```python
twitter_data.columns
```

### Issue #5:
Non-descriptive columns.

#### Define:
The columns (p1, p1_conf,p1_dog, p2, p2_conf, p2_dog, p3, p3_conf, p3_dog) have header names that do not clearly define what information they receive.



#### Code:

```python
twitter_data.rename(columns={'p1' : 'probability_1',
                            'p1_conf' : 'prob_1_confidence',
                            'p1_dog' : 'prob_1_is_dog',
                            'p2' : 'probability_2',
                            'p2_conf' : 'prob_2_confidence', 
                            'p2_dog' : 'prob_2_is_dog', 
                            'p3' : 'probability_3', 
                            'p3_conf' : 'prob_3_confidence', 
                            'p3_dog' : 'prob_3_is_dog'}, inplace=True)
```

#### Test:


```python
twitter_data.columns
```

### Issue #6:
Odd entries in the name column.

#### Define:
In the 'name' column we have entries that could not be dog names.

#### Code:

```python
twitter_data.name = twitter_data.name.str.replace('^[a-z]+', 'None')
```

#### Test:

```python
twitter_data.name.value_counts()
```

### Issue #7:
Four different columns to represent data that can be represented by one main column.

#### Define:
The columns (doggo,floofer,pupper,puppo) can be changed into entries in one main column dog_stage.

#### Code:

```python
twitter_data['dog_stage'] = twitter_data['text'].str.extract('(doggo|floofer|pupper|puppo)')
```

```python
twitter_data.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis = 1, inplace = True)
```

#### Test:

```python
twitter_data.columns
```

```python
twitter_data.dog_stage.value_counts()
```

### Issue #8:
Missing values.

#### Define:
We have 'none' and NaN values in our data. We change the none to NaN.

#### Code:

```python
import numpy as np
```

```python
twitter_data = twitter_data.replace("None", np.nan)
twitter_data = twitter_data.replace("", np.nan)
```

#### Test:

```python
twitter_data
```

### Issue #9:
Data beyond 01/08/2017.

#### Define:
This data is not usable because we are unable to gather the image predictions and we therefore delete it.

#### Code:

```python
twitter_data = twitter_data[(twitter_data.timestamp < '2017/08/02')]
```

#### Test:

```python
twitter_data[twitter_data.timestamp >= '2017/08/02']
```

### Issue #10:
Retweets in my data.

#### Define:
Delete all the retweets.

#### Code:

```python
twitter_data = twitter_data[np.isnan(twitter_data.retweeted_status_id)]
```

#### Test:

```python
twitter_data.info()
```

```python
#Delete the retweeted_status_id column 
twitter_data.drop(['retweeted_status_id'], axis=1, inplace=True)
```

```python
#storing the data
twitter_data.to_csv('twitter_archive_master', index=False)
```
